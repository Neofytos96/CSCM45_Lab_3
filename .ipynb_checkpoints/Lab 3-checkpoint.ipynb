```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
x_values = np.load("x_values.npy")
y_values = np.load("y_values.npy")
```

```python
regression = LinearRegression().fit(x_values, y_values)
```

```python
test_x_values = np.load("test_x_values.npy")
```

```python
y_predictions = regression.predict(test_x_values)
```

```python



plt.scatter(x_values, y_values)
plt.plot(test_x_values , y_predictions)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Observed sample points for Linear Regression')
plt.show(block=False)
```

```python
y_intercept = regression.intercept_
gradient = regression.coef_

test_x = 0.48
# by usign the y = mx+c
predicted_y_value = gradient*test_x+y_intercept
# regression.predict({"y_predictions":new_X})
print(f'Predicted y value for {test_x} is {predicted_y_value}')
```

## Task 3.2

```python
x_data = np.load("wineData.npy")
y_labels = np.load("wineLabels.npy")
print ('x size',x_data.size)
print ('y size',y_labels.size)
```

```python
x_data[0].size
print(13*178)
```

```python
#train - test split 
X_train, X_test, y_train, y_test = train_test_split(x_data, y_labels, test_size=0.2, random_state=100)
y_train.size
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test, marker="x")
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.xlabel('Feature 0')
plt.ylabel('Feature1')
plt.title('Wine Dataset in original feature space')
plt.show(block=False)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_train)
principal_df = pd.DataFrame(data = principal_components
             , columns = ['principal component 1', 'principal component 2'])

finalDf = principal_df
finalDf['Target'] = y_train

finalDf

# pca = PCA(n_components=2)
principal_components = pca.transform(X_test)
principal_df = pd.DataFrame(data = principal_components
             , columns = ['principal component 1', 'principal component 2'])

finalDf_test = principal_df
finalDf_test['Target'] = y_test

finalDf_test
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('Wine Dataset in Principle Component space', fontsize = 20)
targets = [0,1,2]
colors = ['r', 'b', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               ,c=color
               )
    
for target, color in zip(targets,colors):
    indicesToKeep = finalDf_test['Target'] == target
    ax.scatter(finalDf_test.loc[indicesToKeep, 'principal component 1']
               , finalDf_test.loc[indicesToKeep, 'principal component 2']
               ,c=color
               ,marker="x"
            )
ax.legend(targets)
ax.grid()
```

## Task 3.3

```python
lda_model = LDA(n_components=2)
lda_components = lda_model.fit_transform(X_train, y_train)
lda_df = pd.DataFrame(data = lda_components
             , columns = ['LDA component 1', 'LDA component 2'])

finalDf = lda_df
finalDf['Target'] = y_train




lda_components = lda_model.transform(X_test)
lda_df_test = pd.DataFrame(data = lda_components
             , columns = ['LDA component 1', 'LDA component 2'])

finalDf_test = lda_df_test
finalDf_test['Target'] = y_test
# finalDf
finalDf_test
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('LDA Component 1', fontsize = 15)
ax.set_ylabel('LDA Component 2', fontsize = 15)
ax.set_title('Wine Dataset in Linear Discriminant space', fontsize = 20)
targets = [0,1,2]
colors = ['r', 'b', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'LDA component 1']
               , finalDf.loc[indicesToKeep, 'LDA component 2']
               ,c=color
               )
    
for target, color in zip(targets,colors):
    indicesToKeep = finalDf_test['Target'] == target
    ax.scatter(finalDf_test.loc[indicesToKeep, 'LDA component 1']
               , finalDf_test.loc[indicesToKeep, 'LDA component 2']
               ,c=color
               ,marker="x"
            )
ax.legend(targets)
ax.grid()
```

Notice that unlike with PCA, we now need to provide the class labels for our training data (why?)
Because LDA is a supervised machine learning algorithm so it means that it needs the class labels in order to be trained.

```python


n_class = np.max(y_train) + 1
n_samples = X_train.shape[0]
n_features = X_train.shape[1]

mu = np.mean(X_train, axis=0)
mu_class = np.zeros((n_class, n_features))
for i_class in range(n_class):
    mu_class[i_class, :] = np.mean(X_train[y_train==i_class, :], axis=0)
```

```python
from scipy.stats import zscore

# mu_class = X_train - mu_class
# train_mean = zscore(X_train)
train_mean = np.mean(X_train, axis=0)
columns = len(X_train[0])
rows = len(X_train)

# print(columns,"",rows)
mean_centered = np.zeros((rows, columns))
# print(X_train.shape)
for row in range(len(X_train)):
    temp = X_train[row]- train_mean
#     print (temp.shape)
    mean_centered[row]= temp



mean_centered
columns = len(X_test[0])
rows = len(X_test)
# print(columns,"",rows)
mean_centered_test = np.zeros((rows, columns))
# print(X_train.shape)
for row in range(len(X_test)):
    temp = X_test[row]- train_mean
#     print (temp.shape)
    mean_centered_test[row]= temp
    
mean_centered_test
```

```python
u, s,vh = np.linalg.svd(mean_centered)
u_test, s_test,vh_test = np.linalg.svd(mean_centered_test)
```

```python
projection_matrix = vh[0::1]
projected_train_data = mean_centered @ np.transpose(projection_matrix)
```

```python
projected_train_data
```

```python
projection_matrix_test = vh[0::1]
projected_test_data = mean_centered_test @ np.transpose(projection_matrix_test)
projected_train_data.shape
# y_test.shape
```

The slicing into vh selects our top significant eigenvectors, but why do we need the transpose? 
if we have a lot of dimensions then it would take too much time to calculate the principal components

we need to use the same projection matrix from before (why?)
